### Parse offenses

looks like list format but this is actually a string

```python
df.offenses.head()
```

```python
b = df.offenses.str[1:-1].str.split(', ', expand = True).replace("'", "")
```

```python
b['combined'] = b.values.tolist()
b['combined'] 
```

problem here, duplicate columns

```python
parsed = []

for entry in b['combined']:
    #drop duplicates and also check for null values
    c = np.unique([x for x in entry if x != None])
    parsed.append(c)
    
parsed
```
